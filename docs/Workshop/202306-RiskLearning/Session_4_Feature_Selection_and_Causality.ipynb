## Feature Selection

- Switch to the "Conditional Independence" tab.

- Custome the settings:

    - **Initialization**: the initial Markov boundary set, by default, None. You may customize it to one of the feature selections results of the rest 3 methods, i.e., correlation, distance correlation, or feature importance.

    - **p-Threshold**: the significance level of the RCIT test, within 0 to 1. Low thresholds mean strict selection criteria.

    - **Kernel Size**: the number of random Fourier features. Increasing this value can make the results more accurate.

    - **Forward Phase Num**: the number of repetition of the forward selection phase, in the Forward-Backward selection with Early Dropping (FBEDk) algorithm.

- Click on the "UPDATE" button to get the feature selection results based on the updated settings.

    - The figure shows the step-by-step feature selection results of this algorithm. See more details [here](https://selfexplainml.github.io/PiML-Toolbox/_build/html/guides/data/feature_select.html#randomized-conditional-independence-test).

    - The table displays the final results.

- Finally, click on "CONFIRM" button to apply the feature selection for further modeling and analysis.
